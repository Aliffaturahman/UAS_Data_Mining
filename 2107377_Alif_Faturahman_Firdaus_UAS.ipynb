Pilihan dataset yang saya gunakan : jumlah_naik_haji

Menurut saya, jumlah penduduk yang naik haji cocok digunakan dalam kasus IPM karena seperti yang disebutkan PMI sendiri memiliki 3 dimensi

1.   Umum Panjang dan Hidup Sehat

Dengan banyaknya penduduk yang pergi naik haji terlebih lagi kebanyakan orang yang sudah berumur, maka dengan data ini kita bisa mengetahui berapa banyak penduduk berumur tua yang masih sehat, jika tidak sehat maka tidak diperkenankan untuk pergi

2.   Pengetahuan

Dengan banyaknya manfaat naik haji, banyak juga penduduk yang mencari tahu apa, fungsi, tujuan, dan maanfaat dari naik haji itu sendiri

3. Standar Hidup Layak

Dengan mengasumsikan penduduk yang pergi naik haji dengan uangnya sendiri (diluar voucher, sumbangan, atau bantuan orang lain) dengan biaya yang tidak murah. Bisa dikatakan penduduk tersebut memiliki penghasilan lebih karena dapat berpergian ke luar negeri (salah satunya pergi naik haji)

```python
data_haji = '/content/drive/MyDrive/UAS_Data_Mining/Ujian/csv/jumlah_naik_haji.csv'
# data_pmi = '/content/drive/MyDrive/UAS_Data_Mining/Ujian/csv/pmi.csv'

df = pd.read_csv(data_haji)
df.head(10)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.fillna(method='ffill',inplace=True) #sebelumnya
df.fillna(method='bfill',inplace=True) #berikutnya
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df["prov"] = df["prov"].astype("category")
```
